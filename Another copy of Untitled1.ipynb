```python


path = '/content/drive/MyDrive/DATASET/NEWDATASET.csv'

import csv

h, w = [], []

with open(path, 'r') as f:
    r = csv.reader(f)
    next(r)
    for row in r:
        h.append(float(row[1]))
        w.append(float(row[2]))

def calculate_mean(horw):
        total = sum(horw)
        length = len(horw)
        result = total / length
        return result

def calculate_median(horw):
        horw.sort()
        n = len(horw)
        middle = n // 2
        if n % 2 == 0:
            result = (horw[middle - 1] + horw[middle]) / 2
        else:
            result = horw[middle]
        return result

def calculate_mode(horw):
        Counter = {}
        for item in horw:
            if item in Counter:
                Counter[item] += 1
            else:
                Counter[item] = 1
        result = max(Counter, key=Counter.get)
        return result

height_mean = calculate_mean(h)
height_median = calculate_median(h)
height_mode = calculate_mode(h)

weight_mean = calculate_mean(w)
weight_median = calculate_median(w)
weight_mode = calculate_mode(w)

print(f"Height Mean: {height_mean}")
print(f"Height Median: {height_median}")
print(f"Height Mode: {height_mode}")

print(f"Weight Mean: {weight_mean}")
print(f"Weight Median: {weight_median}")
print(f"Weight Mode: {weight_mode}")
```
